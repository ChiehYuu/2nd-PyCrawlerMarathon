## requests + BeautifulSoup


```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup as Soup
url = 'https://www.ptt.cc/bbs/Beauty/M.1556291059.A.75A.html'
cookies = {
    'over18':'1'
}
response = requests.get(url,cookies = cookies)
soup = Soup(response.text)
soup
```

## Grab + PyQuery

```python
'''
Your Code
'''
from grab import Grab
from pyquery import PyQuery as pq
gr = Grab()
response = gr.go('https://www.ptt.cc/bbs/Beauty/M.1556291059.A.75A.html',cookies = cookies)
doc = pq(response.body)
print(doc('title'))
doc.text()
```

1. requests + BeautifulSoup是一次將所有的網頁資料都擷取下來並存在變數中，並且是以html標籤型式的方式儲存成字串。
2. Grab + PyQuery則是同樣一次擷取但是將標籤直接省略並作為查詢用的關鍵字儲存成字串
