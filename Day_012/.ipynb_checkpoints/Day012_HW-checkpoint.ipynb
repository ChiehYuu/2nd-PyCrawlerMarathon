### ① 取出今天所有的發文

```python
'''
Your Code
'''
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
'''
Your Code
'''
import pandas as pd
date_list = []
class_list = []
title_list = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    class_list.append(d.find('em').text)
    date_list.append(d.find('span').text)
    title_list.append(d.find_all('a')[-1].text)
dic = dict(類別=class_list,時間=date_list,標題=title_list)
df = pd.DataFrame(dic)
df
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
cal = df.groupby('類別').count()
max_cal = cal[cal == cal.max()[0]].dropna().index[0]
print('{} 的文章最多，一共有{}篇'.format(max_cal,cal.max()[0]))
```
